```python
import numpy as np
from matplotlib import pyplot as plt
import seaborn as sns
```

```python
C=1
def gap(s, n_l, n_u, d=2):
  return C*np.minimum(np.minimum(n_l/((s**2)*n_u),((s**2)*n_u)/n_l)*d/(n_l +(s**2)*n_u), s)
```

```python
def plot_gap(ax, s, n_l, n_u, d=2, label="$n_l= 10, n_u=10000$", color= 'b', marker="o", markersize=10):
  g = gap(s=s, n_l=n_l, n_u=n_u, d=d)
  ax.plot(s, g, label=label, color=color,linewidth=5, marker=marker, markersize=markersize, markevery=100)
  ax.axvline(x=np.sqrt(n_l/n_u), linestyle= '--', color=color,linewidth=5)

```

```python
s = np.linspace(1e-5, 1, 1000)

sns.set_theme()

fig, ax = plt.subplots(figsize=(10,8))

#plt.figure()
#plot_gap(ax, s, n_l = 10, n_u= 10000, d=5, label="$n_l= 10, n_u=10000$")
plot_gap(ax, s, n_l = 100, n_u= 5000, d=5, label="$n_l= 10$", color='r', marker="o", markersize=15)
plot_gap(ax, s, n_l = 100, n_u= 10000, d=5, label="$n_l= 100$", color='b', marker="*", markersize=20)
plot_gap(ax, s,  n_l = 1000, n_u= 10000, d=5, label="$n_l= 1000$", color='g', marker="P", markersize=20)
sns.despine()

plt.yscale('log')
plt.xlabel("$s$")
plt.ylabel('Gap')
plt.legend(loc='lower right')

sns.set(font_scale=3)
ax.set_title(f"Err(SSL-S) - Err(SSL-W)")
ax.set_xlabel("SNR s", fontsize=35)
ax.set_ylabel("Error gap", fontsize=35)
ax.tick_params(axis="x", labelsize=25, size=30)
ax.tick_params(axis="y", labelsize=25, size=30)
ax.legend(fontsize=30, facecolor="w")
plt.margins(0.03)

ax.set_facecolor('white')
ax.set_axisbelow(True)
plt.grid(color='gray', linestyle='solid')

plt.tight_layout()
plt.savefig(f'/Users/alexandrutifrea/Projects/SSL_lower_bound/SSL_compatibility_lb/neurips2023/figures/sslw_ssls_theory_gap.pdf')


#plt.legend(bbox_to_anchor=(1.0, 1.0))
#ax.legend.set_bbox_to_anchor((.61, .6))
```

```python
def binary_search_smallest(objective, min, max, tol=1):
  while np.abs(min - max) > tol:
    m = int((min + max) / 2)
    if objective(m):
      max = m
    else:
      min = m
  print(max)
  return max
```

```python
c_0 = 1e-5
C_0 = np.log(1/c_0)
C_2 = 2*c_0**2
C_1 = 4*172.05*c_0**2
```

```python
s = np.linspace(0, 0.2, 1000)
lbs_1 = lb(s, n_l = 1, n_u= 10000, d=2)
lbs_100 = lb(s, n_l = 100, n_u= 10000, d=2)
lbs_10 = lb(s, n_l = 10, n_u= 10000, d=2)
lbs_1000 = lb(s, n_l = 1000, n_u= 10000, d=2)
lbs_10000 = lb(s, n_l = 10000, n_u= 10000, d=2)
lbs_1000000 = lb(s, n_l = 1000000, n_u= 10000, d=2)
```

```python
def lb(s, n_l, n_u, d=2):
  return 2* np.sqrt(C_0)*(C_0*d -1)/(3*np.sqrt(3)*C_0*d)*np.sqrt((d-1/C_0)/(C_2*n_l + C_1*(s**2)*n_u))
```

```python
plt.figure()
plt.plot(s, lbs_1, label="n_l= 1")
plt.plot(s, lbs_10, label="n_l= 10")
plt.plot(s, lbs_100, label="n_l= 100")
plt.plot(s, lbs_1000, label="n_l= 1000")
plt.plot(s, lbs_10000, label="n_l= 10000")
plt.plot(s, lbs_1000000, label="n_l= 1000000")

plt.yscale('log')
plt.xlabel('snr')
plt.ylabel('estimation error(up to constants)')
plt.legend()
```
